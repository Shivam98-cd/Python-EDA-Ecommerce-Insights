```python
#Section 4: Bucketing using Function and Apply
"""Define a function called amount_bucket that returns:
– High if Quantity * Price >= 5000
– Medium if between 2000 and 4999
– Low if below 2000"""
df["sales"]=(df["quantity"]*df["price"])
def amount_bucket(x):
    
    if (x>=5000):
        return "High"
    elif (x>=2000 and x<4999):
        return "Medium"
    else:
        return "Low"
df["amount_bucket"]=df["sales"].apply(amount_bucket)
df.head().round()
    
    
```

```python
#8. Create a new column called Amount.
df=df.rename(columns={"price":"Amount"})
```

```python
#23. Create a new column called Discount_Flag with the following logic:
#If Price > 1500 then Yes else No.
def Discount_Flag(x):
    if x>1500:
        return "Yes"
    else:
        return "No"
df["Discount_Flag"]=df["Amount"].apply(Discount_Flag)
df.head()
```

```python
#24. Create a new column called High_Quantity_Flag where Quantity > 4 then Mark High else Normal.
def High_Quantity_Flag(x):
    if x>4:
        return "High"
    else:
        return "Normal"
df["High_Quantity_Flag"]=df["quantity"].apply(High_Quantity_Flag)
df.head()
```

```python
#25. Create a Profit column using if logic:
#If Product_Category is Electronics profit = Amount * 0.12 else Amount * 0.08.
def Profit(row):
    if row["product_category"] == "Electronics":
        return row["Amount"] * 0.12
    else:
        return row["Amount"] * 0.08

df["Profit"] = df.apply(Profit, axis=1)
```
